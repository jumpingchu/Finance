```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader as web
```

```python
df = web.DataReader("AMZN", "yahoo", start = "20150101", end = "2018-12-22")
# web.DataReader(['name', 'data_source=None', 'start=None', 'end=None')
# 日期形式可以用20100101 or 2010/1/1 or 2010.1.1都可以
```

```python
df.head()
```

```python
p = df["Adj Close"]
# 僅抓取調整收盤價
```

```python
p.shape
```

```python
p.plot()
```

# 報酬率

```python
r = p.diff()/p
# diff = difference
# 除上p本身後，代表著報酬率 r
```

```python
r.head()
```

```python
r[-10:].plot()
```

# 移動平均線

```python
p.head()
```

```python
p.plot()
p.rolling(window=20).mean().plot()
p.rolling(window=60).mean().plot()
p.rolling(window=120).mean().plot()
p.rolling(window=240).mean().plot()
```
